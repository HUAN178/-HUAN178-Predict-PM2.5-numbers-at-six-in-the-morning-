```python
trains = train.groupby("測項") 
train_PM25 = trains.get_group("PM2.5").reset_index(drop=True)
train_PM25
```

```python
tests = test.groupby("測項")
test_PM25 = tests.get_group("PM2.5").reset_index(drop=True)
test_PM25
```

```python
X_train = train_PM25[['00','01','02','03','04','05']]
X_test = test_PM25[['00','01','02','03','04','05']]
y_train = train_PM25['06']
y_test =test_PM25['06']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
from sklearn import datasets, linear_model

regr = linear_model.LinearRegression()

# 訓練模型
regr.fit(X_train_std, y_train)
print("MSE: %.2f" % np.mean((regr.predict(X_test_std) - y_test) ** 2))
print("R Square_test:",regr.score(X_test_std, y_test))
print("R Square_train:",regr.score(X_train_std, y_train))
```

```python
from sklearn.ensemble import RandomForestClassifier # 隨機森林
RFC = RandomForestClassifier(random_state=1000,n_estimators=100,min_samples_split=20,oob_score=True)
RFC.fit(X_train, y_train)
print("MSE: %.2f" % np.mean((RFC.predict(X_test) - y_test) ** 2))
print("R Square_test:",RFC.score(X_test, y_test))
```
